```python
#program to solve quadratic equation
a = float(input("Enter coefficeant of x^2 : "))
b = float(input("Enter coefficient of x : "))
c = float(input("Constant value c : "))
dis = (b**2) - (4*a*c)
if(dis < 0):
  print("No real roots for the equation.")
else:
  x1 = (-b + (dis**0.5))/(2*a)
  x2 = (-b - (dis**0.5))/(2*a)
  print(f"The roots of equation are {x1} and {x2}")
```

```python
#armstrong number
#a number is armstrong if sum of its digits raised to power of no. of digits

arm = int(input("Enter the number to check : "))
arm_digit = len(str(arm))
temp = arm
sum_of_powers = 0

while temp > 0:
  digit = temp%10
  sum_of_powers += digit ** arm_digit
  temp //= 10

if(sum_of_powers == arm ):
  print('Its is an armstrong number')
else:
  print('It is not an armstrong number')
```

**RANDOM module**

Using random.seed()

```python
import random
for i in range(2):
    print(random.randint(1, 1000))

for i in range(2):
    random.seed(1) #
    print(random.randint(1, 1000))
```

*Notice that generating random numbers without using the .seed() method results in different outputs on each run, whereas using it ensures consistent results every time.*
