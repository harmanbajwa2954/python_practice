**Q1. Simple Interest**

```python
p = float(input("Enter the Principle Amount : "))
r = float(input("Enter rate of interest : "))
t = float(input("Enter the duration (in Years) : "))
si = (p*r*t)/100
print(f"The Simple Interest after {t} years with rate {r}% is {si}")
```

**Q2. Compound Interest**

```python
p = float(input("Enter the Principle Amount : "))
r = float(input("Enter the rate of Interest : "))
t = float(input("Enter the duration (in Years) : "))

amount = p*((1+r/100)**t)

ci = amount-p

print(f"The Total amount after {t} years with {r}% of interest is {amount}.")
print(f"The Compound Interest will be {ci}")
```

**Q1. Print all prime numbers between 1 to N**

```python
n = int(input("Enter the number : "))
for num in range(0, n):
    if num >= 1:
        for i in range(2, num):
            if (num % i) == 0:
                break
        else:
            print(num, end=" ")
```

**Q2. Print the sum of all prime numbers between A to B.**

```python
A = int(input("Enter the first number : "))
B = int(input("Enter second number : "))
sum = 0
for num in range(A, B):
    if num >= 1:
        for i in range(2, num):
            if (num % i) == 0:
                break
        else:
            sum += num

print(f"The sum of all prime numbers between {A} and {B} is {sum}")
```

**Q5. Check if a string is a palindrome using loop.**

```python
s = input("Enter the string to check : ")
j = -1
flag = 0
for i in range(len(s)):
  if(s[i] != s[j]):
    flag = 1
    break
  j = j-1

if(flag == 1):
  print("The string is not a palindrome.")
else:
  print("The string is a pallindrome.")

```

**Q6. Find sum of digits until the result becomes a single digit (e.g., 987 →
9+8+7 = 24 → 2+4 = 6).**

```python
n = int(input("Enter number: "))

while n >= 10:
    p = 0
    temp = n
    while temp > 0:
        p += temp % 10
        temp //= 10
    n = p
    print(n)

print("Single digit result:", n)
```

**Q7. Check if a number is an Armstrong number.**

```python
arm = int(input("Enter the number to check : "))
arm_digit = len(str(arm))
temp = arm
sum_of_powers = 0

while temp > 0:
  digit = temp%10
  sum_of_powers += digit ** arm_digit
  temp //= 10

if(sum_of_powers == arm ):
  print('Its is an armstrong number')
else:
  print('It is not an armstrong number')
```

**Q12. Reverse a number (e.g., 123 → 321).**

```python
term = int(input("Enter the number to reverse : "))
reverse = 0
while term > 0:
  temp = term%10
  reverse = reverse*10 + temp
  term //= 10
print(f"The reverse number is {reverse}")
```

**Q13.For numbers 1 to N:**
