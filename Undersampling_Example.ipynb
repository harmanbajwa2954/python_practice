```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from imblearn.under_sampling import RandomUnderSampler
```

```python
df = pd.read_csv('imbalanced_data.csv')
df
```

```python
sns.countplot(x='Target', data=df)
plt.title("Original Data")
plt.show()
```

```python
X = df[['Feature_1', 'Feature_2']]
y = df['Target']
```

```python
# Perform Random Undersampling
undersampler = RandomUnderSampler(random_state=42)
X_under, y_under = undersampler.fit_resample(X, y)
```

```python
sns.countplot(x=y_under, data=X_under)
plt.title("Undersampled Data")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

#original data
ax1.scatter(X[y==0]['Feature_1'], X[y==0]['Feature_2'], label='Majority (0)', alpha=0.6)
ax1.scatter(X[y==1]['Feature_1'], X[y==1]['Feature_2'], label='Minority (1)', color='red')
ax1.set_title(f'Original Imbalanced Data\n{y.value_counts().to_dict()}')
ax1.legend()

#undersampled data
ax2.scatter(X_under[y_under==0]['Feature_1'], X_under[y_under==0]['Feature_2'], label='Majority (0)', alpha=0.6)
ax2.scatter(X_under[y_under==1]['Feature_1'], X_under[y_under==1]['Feature_2'], label='Minority (1)', color='red')
ax2.set_title(f'After Undersampling\n{y_under.value_counts().to_dict()}')
ax2.legend()

plt.tight_layout()
plt.show()
```
